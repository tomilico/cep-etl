```python
import pandas as pd
import sqlalchemy
import os
from pathlib import Path
```

```python
current_dir = Path(os.getcwd())
file_path = current_dir / '3-gold-enriched' / 'query.sql'
```

```python
# definir a string de conexao com PostgreSQL
db_url = "postgresql://postgres:postgres@localhost:5432/postgres"
db_path = db_url  #p manter compatibilidade com o resto do codigo
```

```python
with open('3-gold-enriched/query.sql', 'r', encoding='utf-8') as file:
    query = file.read()

#conectar ao banco de dados e executar a query
with sqlalchemy.create_engine(db_path).connect() as conn:
    df = pd.read_sql_query(query, conn)

#vizualizacao dos dados
df.head()
```
